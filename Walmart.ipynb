```python
ph_maxsale=ph_data[['Store','Weekly_Sales']]
ph_maxsale.Weekly_Sales.max()
```

```python
store=ph_maxsale.groupby(ph_maxsale.Store)
maxstore=ph_maxsale.groupby(ph_maxsale.Store).sum()
maxstore.sort_values(by='Weekly_Sales',ascending=False,inplace=True)
maxstore
```

```python
std=store.std()

std.sort_values(by='Weekly_Sales',ascending=False,inplace=True)
std
```

```python
co=store.std()/store.mean()
co.sort_values(by='Weekly_Sales',ascending=False,inplace=True)
co
```

```python
q=ph_data[['Store','Weekly_Sales','Date']]
q['Date'] = pd.to_datetime(q['Date'])
start_date_1 = "2012-04-01"
end_date_1 = "2012-06-30"
after_start_date_1 = q["Date"] >= start_date_1
before_end_date_1 = q["Date"] <= end_date_1
between_two_dates_1 = after_start_date_1 & before_end_date_1
a_1=q.loc[between_two_dates_1]
a_1.sort_values(by='Date',ascending=True,inplace=True)
b_1=a_1.groupby(a_1.Store).sum()

b_1
```
